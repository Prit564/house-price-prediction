## House Price Prediction - Univariate Analysis

Objective: To understand the distribution, spread, and basic characteristics of each feature in the dataset.

```python
import pandas as pd
df = pd.read_csv(r"clean_data.csv")
df.head()
```

```python
df.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.histplot(df['area_sqft'],bins=20,kde=True)
plt.title('Property area')
```

### Zooming in the dense part:

```python
df1 = df[df['area_sqft'] < 4000]
sns.histplot(df1['area_sqft'],bins=20,kde=True)
plt.title('Property area')
```

The initial histogram reveals a severe right-skew in property sizes, driven by extreme outliers approaching 10,000 square feet that visually compress the bulk of our data. By applying a filter to exclude properties over 4,000 square feet, we successfully isolate the core market of standard residential homes. This adjustment exposes a much healthier, nearly normal distribution centered around 2,200 square feet.

```python
sns.countplot(x=df['rooms'])
plt.title('Number of rooms')
```

Apartments come up of varying number of rooms from small apartments of 2 rooms to huge 7 room properties. However 4 rooms are pretty standard and most common if we choose a random apartment.

```python
plt.figure(figsize=(12, 6))
sns.histplot(data=df, x='build_year', binwidth=5, color='coral', edgecolor='black')

plt.title('Property Distribution by Build Year', fontsize=14, pad=15)
plt.xlabel('Build Year', fontsize=12)
plt.ylabel('Count of Properties', fontsize=12)

plt.show()
```

Properties were built in a pretty cosistent frequency over the years. However, a spike in the count is seen over the final 10 years. This means, the datasheet might contain maximum of recent building records rather than past ones or simply the property building frequency has increased continually in the area.

```python
df2 = df[df['price'] < 1000000]
sns.histplot(df1['price'],bins=20,kde=True)
plt.title('Property area')
```

Most of the houses price ranges between 250000 to 1000000. This is the moderate range. In this range 550000 to 650000 priced properties are most common. Apart from these, there are few in number but properties of price 1000000 to 2000000 available.

```python
sns.countplot(data=df, x='furnishing', hue='furnishing', palette='Set2')
plt.title('Furnishing Type')
```

An analysis of property distribution reveals that semi-furnished units constitute the clear majority of the market. While fully furnished properties also maintain a substantial presence, unfurnished options represent the least prevalent segment in the dataset.

```python
sns.countplot(data=df, x='property_type', hue='property_type', palette='Set2')
plt.title('Furnishing Type')
```

Apartments constitute the vast majority of property types within the dataset, followed by independent houses as the secondary market category. Conversely, duplexes and villas represent the least prevalent segments, accounting for the lowest volume of listings.

## Summary:

* **Property Size:** The core market consists of standard homes centered around 2,200 square feet. Extreme outliers (up to 10,000 sq ft) skew the overall data, but filtering out properties over 4,000 sq ft reveals a healthy, normal distribution.
* **Room Count:** While available properties range from 2 to 7 rooms, 4-room layouts are the established market standard.
* **Age of Properties:** Construction history has been largely consistent, but there is a distinct spike in properties built within the last decade. This suggests either a recent local construction boom or a dataset that is heavily weighted toward newer builds.
* **Pricing:** Most properties fall into a moderate price bracket of 250,000 to 1,000,000, with the highest concentration clustered between 550,000 and 650,000. A smaller, premium segment of the market spans from 1,000,000 to 2,000,000.
* **Furnishing:** Semi-furnished homes dominate the market, closely followed by fully furnished options. Unfurnished properties are relatively rare.
* **Property Type:** Apartments are by far the most common property type, with independent houses making up the second largest category. High-end variations like duplexes and villas account for the smallest share of the market.
