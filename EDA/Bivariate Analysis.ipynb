## House Price Prediction - Bivariate Analysis

**Objective:** Analyzing different variables with respect to "price" to understand how they are related to each other and providing insights.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"clean_data.csv")
```

```python
df.head()
```

### Let's analyze how the property area is related to the prices first

```python
df.plot.scatter(x='area_sqft', y='price', alpha=0.5, figsize=(10, 6))

plt.title('Property Price vs. Area')
plt.xlabel('Property Area (sq ft)')
plt.ylabel('Price')
plt.show()
```

### Just zooming in the dense portion:

```python
df.plot.scatter(x='area_sqft', y='price', alpha=0.5, figsize=(10, 6))
plt.xlim(0, 4000)
plt.title('Property Price vs. Area(Close View)')
plt.xlabel('Property Area (sq ft)')
plt.ylabel('Price')
plt.tight_layout()
plt.show()
```

The graph reveals some important insights about property area vs price data.
1. First of all, it is evident that maximum of data cluster from area of 500-4000 sqft range meaning most of the properties bear an area of this amount.
2. There is a pretty linear trend in the graph meaning prices increase as per area of property increases.
3. One of the most important trend to see is, for a single area sqft value (say 2000 sqft) there are numerous vertical line of points in the y axis (price data). This implies that same area property may come up with different prices according to facilities.
4. This variation is more common for data from about 1600-2600 sqft.

```python
plt.figure(figsize=(8, 8))
df1=df[df['price']<1000000]
sns.boxplot(data=df1,y='price',x='rooms',color='tab:blue')
plt.title('Price vs rooms number')
plt.xlabel('Rooms')
plt.tight_layout()
```

This graph shows the price and room number dependence. It is visible that it follows a linear trend. As room number increases, property price also gets increased. One thing to mention that I have cut the outlier data as they interfere with the presence of the graph.

```python
plt.figure(figsize=(8, 8))
df1=df[df['price']<1000000]
sns.boxplot(data=df1,y='price',x='property_type',color='tab:blue')
plt.title('Price vs property type')
plt.xlabel('Property')
plt.tight_layout()
```

Ranges of price for each type of property is pretty similar, however the trend is to notice. Apartments are so far the cheapest of the properties. Duplex and Independent houses share a much similar range placing both of them at the 2nd place in the list. The most expensive properties are the 'Villas'

```python
plt.figure(figsize=(8, 8))
df1=df[df['price']<1000000]
sns.boxplot(data=df1,y='price',x='furnishing',color='tab:blue')
plt.title('Price vs Furnishing type')
plt.xlabel('Furnishing')
plt.tight_layout()
```

As expected, furnishing a property shifts the entire price distribution upward. While semi-furnished and unfurnished homes have similar average values, semi-furnished homes have a higher ceiling for luxury outliers, and fully furnished homes command a clear price premium across the board.

```python
plt.figure(figsize=(8, 8))
df1=df[df['price']<1000000]
sns.boxplot(data=df1,y='price',x='street_type',color='tab:blue')
plt.title('Price vs Street type')
plt.xlabel('Street')
plt.tight_layout()
```

Location type acts as a strong price escalator; gated societies command a clear premium floor for luxury buyers, residential lanes offer the most affordable entry points, and mid-tier street types (main roads, corner plots, highway facing) share a highly competitive baseline price range.
